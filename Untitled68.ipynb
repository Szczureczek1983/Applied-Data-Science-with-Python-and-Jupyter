```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.cluster import KMeans

target_column = 'medium'
features = all_data.drop('medium', axis = 1).values
target = all_data[target_column]

km = KMeans(n_clusters = 3, init= 'random',
             n_init = 10,
             max_iter = 300,
             random_state = 100)
y_km = km.fit_predict(features)
```

```python
plt.scatter(features[y_km == 0,0],
            features[y_km == 0,1],
            s=50, c='lightgreen', marker="o", edgecolor = "black",
            label = 'Cluster 1')
plt.scatter(features[y_km == 1,0],
            features[y_km == 1,1],
            s=50, c='blue', marker="s", edgecolor = "black",
            label = 'Cluster 2')
plt.scatter(features[y_km == 2,0],
            features[y_km == 2,1],
            s=50, c='orange', marker="v", edgecolor = "black",
            label = 'Cluster 3')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s = 250, marker = "*", c = "black",
            label = "Centroids")
plt.legend(scatterpoints = 1)
plt.grid()
plt.show()
```

```python

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.25, random_state= 42)
scaler = StandardScaler()
scaled_train_X = scaler.fit_transform(X_train)
scaled_test_X = scaler.fit_transform(X_test)

```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=3, random_state = 42)
tree.fit(features, target)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
clf = GaussianNB()
clf.fit(scaled_train_X, y_train)
y_pred = clf.predict(scaled_test_X)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(3)
knn.fit(scaled_train_X, y_train)
y_pred = knn.predict(scaled_test_X)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
svc = SVC(kernel = 'rbf', random_state = 42)
scores = cross_val_score(svc, scaled_train_X, y_train, cv = 3, scoring = 'accuracy')
print("SVC with rbf kernel -> cross validation accuracy: mean = %0.2f, std = %0.2f" %(np.mean(scores), np.std(scores)))
```

```python
from sklearn.ensemble import AdaBoostClassifier
Ada = AdaBoostClassifier(n_estimators=1000, random_state=42)
score = cross_val_score(Ada, scaled_train_X, y_train, cv=3, scoring = 'accuracy', n_jobs = 5)
print(" cross validation accuracy: mean = %0.2f, std = %0.2f" %(np.mean(score), np.std(score)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
model = RandomForestClassifier(n_estimators= 1000)
model.fit(scaled_train_X, y_train)
y_pred = model.predict(scaled_test_X)
print(metrics.classification_report(y_pred, y_test))
```

```python
import numpy as np
import tensorflow as tf
imput_size = 30
output_size = 3
hidden_layers = 200
model = tf.keras.Sequential([tf.keras.layers.Dense(hidden_layers, activation = 'relu'),
                             tf.keras.layers.Dense(hidden_layers, activation = 'relu'),
                             tf.keras.layers.Dense(output_size, activation = 'softmax')])
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics =['accuracy'])
model.fit(scaled_train_X, y_train, epochs = 100, verbose = 2)
```

```python
score = model.evaluate(scaled_test_X, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
